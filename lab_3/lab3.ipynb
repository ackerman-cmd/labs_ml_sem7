С возрастом явно выброс, клиент не может быть по логике старше 100 лет и поскольку данных у нас немного, применим винзоризвацию

Всё, что ниже 1-го перцентиля (1%), заменяется на значение 1-го перцентиля.

Всё, что выше 99-го перцентиля (99%), заменяется на значение 99-го перцентиля.


```python
def winsorize(df, cols, q_low=0.01, q_high=0.99):
    for col in cols:
        low, high = df[col].quantile([q_low, q_high])
        df[col] = df[col].clip(lower=low, upper=high)
    return df

df = winsorize(df, ["Income", "Age"] + money_cols)
```

Ожидаем что precision + recall будут невысокие(сильный дисбаланс), но главное чтобы roc-auc был неплохой и далее будем повышать f1

```python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score

def print_classification_metrics(y_true, y_pred, y_proba=None):
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    print(f"Precision: {precision:.4f}")
    print(f"Recall:    {recall:.4f}")
    print(f"F1-score:  {f1:.4f}")

    if y_proba is not None:
        roc_auc = roc_auc_score(y_true, y_proba)
        print(f"ROC-AUC:   {roc_auc:.4f}")
```

### Собственная имплиментация решающего дерева(классификатор)

```python
import numpy as np
from sklearn.base import BaseEstimator, ClassifierMixin

class MyDecisionTreeClassifier(BaseEstimator, ClassifierMixin):
    class Node:
        def __init__(self, gini, num_samples, num_samples_per_class, predicted_class, proba,
                     feature_index=None, threshold=None, left=None, right=None):
            self.gini = gini
            self.num_samples = num_samples
            self.num_samples_per_class = num_samples_per_class
            self.predicted_class = predicted_class
            self.proba = proba
            self.feature_index = feature_index
            self.threshold = threshold
            self.left = left
            self.right = right

    def __init__(self, max_depth=None, min_samples_split=2, min_samples_leaf=1):
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.min_samples_leaf = min_samples_leaf
        self.n_classes_ = None
        self.n_features_ = None
        self.classes_ = None
        self.tree_ = None

    def _to_dense(self, X):
        if hasattr(X, "toarray"):
            X = X.toarray()
        return np.asarray(X, dtype=float)

    def fit(self, X, y):
        X = self._to_dense(X)
        y = np.asarray(y)
        self.classes_ = np.unique(y)
        self.n_classes_ = self.classes_.shape[0]
        self.n_features_ = X.shape[1]
        self.tree_ = self._grow_tree(X, y, depth=0)
        return self

    def _gini(self, y):
        m = y.size
        if m == 0:
            return 0.0
        counts = np.array([np.sum(y == c) for c in self.classes_], dtype=float)
        p = counts / m
        return 1.0 - np.sum(p ** 2)

    def _best_split(self, X, y):
        m, n = X.shape
        if m < self.min_samples_split:
            return None, None

        best_feature, best_threshold = None, None
        best_gini = 1.0

        for feature_idx in range(n):
            x_col = X[:, feature_idx]
            sorted_idx = np.argsort(x_col)
            x_sorted = x_col[sorted_idx]
            y_sorted = y[sorted_idx]

            unique_vals = np.unique(x_sorted)
            if unique_vals.size == 1:
                continue

            thresholds = (unique_vals[:-1] + unique_vals[1:]) / 2.0

            for thr in thresholds:
                left_mask = x_sorted <= thr
                right_mask = ~left_mask

                if left_mask.sum() < self.min_samples_leaf or right_mask.sum() < self.min_samples_leaf:
                    continue

                y_left = y_sorted[left_mask]
                y_right = y_sorted[right_mask]

                g_left = self._gini(y_left)
                g_right = self._gini(y_right)
                g = (y_left.size * g_left + y_right.size * g_right) / m

                if g < best_gini:
                    best_gini = g
                    best_feature = feature_idx
                    best_threshold = thr

        return best_feature, best_threshold

    def _grow_tree(self, X, y, depth):
        num_samples = y.size
        num_samples_per_class = np.array([np.sum(y == c) for c in self.classes_], dtype=float)
        predicted_class_idx = np.argmax(num_samples_per_class)
        predicted_class = self.classes_[predicted_class_idx]
        proba = num_samples_per_class / num_samples if num_samples > 0 else np.zeros_like(num_samples_per_class)
        node_gini = self._gini(y)

        if (self.max_depth is not None and depth >= self.max_depth) or \
           num_samples < self.min_samples_split or \
           node_gini == 0.0:
            return self.Node(node_gini, num_samples, num_samples_per_class, predicted_class, proba)

        feature_idx, threshold = self._best_split(X, y)
        if feature_idx is None:
            return self.Node(node_gini, num_samples, num_samples_per_class, predicted_class, proba)

        left_mask = X[:, feature_idx] <= threshold
        right_mask = ~left_mask
        X_left, y_left = X[left_mask], y[left_mask]
        X_right, y_right = X[right_mask], y[right_mask]

        if X_left.shape[0] == 0 or X_right.shape[0] == 0:
            return self.Node(node_gini, num_samples, num_samples_per_class, predicted_class, proba)

        left_child = self._grow_tree(X_left, y_left, depth + 1)
        right_child = self._grow_tree(X_right, y_right, depth + 1)

        return self.Node(node_gini, num_samples, num_samples_per_class, predicted_class, proba,
                         feature_index=feature_idx, threshold=threshold,
                         left=left_child, right=right_child)

    def _predict_one(self, x):
        node = self.tree_
        while node.left is not None and node.right is not None:
            if x[node.feature_index] <= node.threshold:
                node = node.left
            else:
                node = node.right
        return node

    def predict(self, X):
        X = self._to_dense(X)
        nodes = [self._predict_one(x) for x in X]
        preds_idx = [np.argmax(node.proba) for node in nodes]
        return self.classes_[preds_idx]

    def predict_proba(self, X):
        X = self._to_dense(X)
        nodes = [self._predict_one(x) for x in X]
        return np.vstack([node.proba for node in nodes])
```

В качестве метрик будем смотреть на rmse, r2 и mae 


rmse поскольку штрафует большие ошибки, это то что нам надо, так как цены могут иметь разный порядок. 

r2 чтобы следить какой % разброса таргета обьясняет модель

mae для наглядности(средняя ошибка в долларах)


Если переводить на бизнес-язык: 

RMSE	точность в абсолютных деньгах	

MAE	однозначное понимание среднего отклонения, удобно для бизнеса и заказчика

r2	- то насколько наша модель лучше среднего 


```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def print_regression_metrics(y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    print(f"RMSE: {rmse:.4f}")
    print(f"MAE:  {mae:.4f}")
    print(f"R²:   {r2:.4f}")
    
```

### Собственная имплементация решающего дерева(регрессор)

```python
import numpy as np
from sklearn.base import BaseEstimator, RegressorMixin

class MyDecisionTreeRegressor(BaseEstimator, RegressorMixin):
    class Node:
        def __init__(self, mse, num_samples, value,
                     feature_index=None, threshold=None,
                     left=None, right=None):
            self.mse = mse
            self.num_samples = num_samples
            self.value = value
            self.feature_index = feature_index
            self.threshold = threshold
            self.left = left
            self.right = right

    def __init__(self, max_depth=None, min_samples_split=2, min_samples_leaf=1):
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.min_samples_leaf = min_samples_leaf
        self.n_features_ = None
        self.tree_ = None

    def _to_dense(self, X):
        if hasattr(X, "toarray"):
            X = X.toarray()
        return np.asarray(X, dtype=float)

    def fit(self, X, y):
        X = self._to_dense(X)
        y = np.asarray(y, dtype=float).ravel()
        self.n_features_ = X.shape[1]
        self.tree_ = self._grow_tree(X, y, depth=0)
        return self

    def _mse(self, y):
        if y.size == 0:
            return 0.0
        m = y.size
        mean = y.mean()
        return np.sum((y - mean) ** 2) / m

    def _best_split(self, X, y):
        m, n = X.shape
        if m < self.min_samples_split:
            return None, None

        best_feature, best_threshold = None, None
        best_mse = self._mse(y)

        for feature_idx in range(n):
            x_col = X[:, feature_idx]
            sorted_idx = np.argsort(x_col)
            x_sorted = x_col[sorted_idx]
            y_sorted = y[sorted_idx]

            unique_vals = np.unique(x_sorted)
            if unique_vals.size == 1:
                continue

            thresholds = (unique_vals[:-1] + unique_vals[1:]) / 2.0

            for thr in thresholds:
                left_mask = x_sorted <= thr
                right_mask = ~left_mask

                if left_mask.sum() < self.min_samples_leaf or right_mask.sum() < self.min_samples_leaf:
                    continue

                y_left = y_sorted[left_mask]
                y_right = y_sorted[right_mask]

                mse_left = self._mse(y_left)
                mse_right = self._mse(y_right)
                mse = (y_left.size * mse_left + y_right.size * mse_right) / m

                if mse < best_mse:
                    best_mse = mse
                    best_feature = feature_idx
                    best_threshold = thr

        if best_feature is None:
            return None, None

        return best_feature, best_threshold

    def _grow_tree(self, X, y, depth):
        num_samples = y.size
        value = y.mean() if num_samples > 0 else 0.0
        node_mse = self._mse(y)

        if (self.max_depth is not None and depth >= self.max_depth) or \
           num_samples < self.min_samples_split or \
           node_mse == 0.0:
            return self.Node(node_mse, num_samples, value)

        feature_idx, threshold = self._best_split(X, y)
        if feature_idx is None:
            return self.Node(node_mse, num_samples, value)

        left_mask = X[:, feature_idx] <= threshold
        right_mask = ~left_mask
        X_left, y_left = X[left_mask], y[left_mask]
        X_right, y_right = X[right_mask], y[right_mask]

        if X_left.shape[0] == 0 or X_right.shape[0] == 0:
            return self.Node(node_mse, num_samples, value)

        left_child = self._grow_tree(X_left, y_left, depth + 1)
        right_child = self._grow_tree(X_right, y_right, depth + 1)

        return self.Node(node_mse, num_samples, value,
                         feature_index=feature_idx,
                         threshold=threshold,
                         left=left_child,
                         right=right_child)

    def _predict_one(self, x):
        node = self.tree_
        while node.left is not None and node.right is not None:
            if x[node.feature_index] <= node.threshold:
                node = node.left
            else:
                node = node.right
        return node.value

    def predict(self, X):
        X = self._to_dense(X)
        return np.array([self._predict_one(x) for x in X], dtype=float)
```
